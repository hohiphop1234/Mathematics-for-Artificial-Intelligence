### Linear model

```python
X_bias = np.c_[np.ones(X.shape[0]), X]
```

```python
print(X_bias[:])
```

```python
num_params = len(X_bias[0])
print(num_params)
```

```python
theta = np.random.rand(num_params, 1) * 0.01
print(theta.shape)
```

```python
def predict(X, theta):
    return X.dot(theta)

def compute_loss(y, y_hat):
    N = y.shape[0]
    return np.sum((y - y_hat) ** 2) / N

def compute_gradient(y_hat, y, X):
    N = y.shape[0]
    k = 2 * (y_hat - y)

    return X.T.dot(k) / N

def update_gradient (theta, grad, lr = 0.001):
    theta = theta - lr * grad
    return theta

pred = predict(X_bias, theta)
loss = compute_loss(y, pred)
grad = compute_gradient(pred, y, X_bias)



print(pred)
print(loss)
print(grad)
print(f"Previous theta: {theta[:5]}")
new_theta = update_gradient(theta, grad)
print(f"New theta: {new_theta}")
```
