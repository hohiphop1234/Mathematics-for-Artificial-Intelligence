```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("../data/processed_spam.csv")
```

```python
documents = dataset['processed_text'].tolist()
documents[:5]
```

### Build Vocabulary

```python
vocab = {}
for doc in documents:
    doc = str(doc)
    words = doc.split()
    for word in words:
        if word not in vocab:
            vocab[word] = len(vocab)

print(len(vocab))
```

```python
num_docs = len(documents)
vocab_size = len(vocab)
```

### Bag Of Word

```python
bow_matrix = np.zeros((num_docs, vocab_size), dtype=int)
```

```python
for i, doc in enumerate(documents):
    doc = str(doc)
    for word in doc.split():
        if word in vocab:
            word_index = vocab[word]
            bow_matrix[i, word_index] += 1
```

```python
print(bow_matrix[:2])
```

### TF-IDF

```python
df = np.sum(bow_matrix > 0, axis = 0)
print(df[:10])
```

```python
idf_vector = np.log((1 + num_docs) / (1 + df)) + 1
print(idf_vector[:10])
```

```python
tfidf_matrix = bow_matrix * idf_vector
print(np.round(tfidf_matrix[0], 2))
```
